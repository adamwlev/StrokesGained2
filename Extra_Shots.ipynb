```python
def read_file_and_see_how_many_duplicates(year):
    data = pd.read_csv('data/rawdata/' + str(year) + '.txt', sep = ';')
    data.columns = np.array([str(i).strip() for i in list(data.columns.values)])
    print data.columns
    tot_size = float(len(data))
    sus_size = float(len(data[data['Shot']>data['Hole Score']]))
    pct = sus_size/tot_size*100
    print 'Year: %d' % year
    print 'Total Shots: %d' % len(data)
    print 'Suspect Shots: %d' % len(data[data['Shot']>data['Hole Score']])
    print 'Suspect Shots as percent of Total: %g' % pct
```

```python
## let's see if for any of the suspect shots (shot>hole score), the hole score is wrong, rather than there being extra
## shots in the data set. Will select 50 randomly chosen shots from across all years and check the hole score manually
## with other sources

def make_df_of_suspects():
    data = pd.read_csv('data/rawdata/2016.txt', sep = ';')
    data.columns = np.array([str(i).strip().replace(' ','_') for i in list(data.columns.values)])
    df = pd.DataFrame(data.iloc[[1,2]])
    for y in range(2003,2017):
        data = pd.read_csv('data/rawdata/' + str(y) + '.txt', sep = ';')
        data.columns = np.array([str(i).strip().replace(' ','_') for i in list(data.columns.values)])
        df = df.append(data[data.Shot>data.Hole_Score])
    df.drop(df.head(2).index, inplace=True)
    return df
```

```python
def holes_with_suspect_shots():
    dfs = []
    for y in range(2003,2017):
        data = pd.read_csv('data/rawdata/' + str(y) + '.txt', sep = ';')
        data.columns = np.array([str(i).strip().replace(' ','_') for i in list(data.columns.values)])
        suspect_subset = data[data.Shot>data.Hole_Score]
        suspect_ids = []
        rand_indices = [random.choice(range(len(suspect_subset))) for _ in range(15)]
        for row in rand_indices:
            suspect_ids.append(suspect_subset.iloc[row][['Tournament_Name','Player_First_Name','Player_Last_Name','Round','Hole']].tolist())
        for ID in suspect_ids:
            t_name = ID[0]
            f_name = ID[1]
            l_name = ID[2]
            ROUND = ID[3]
            hole = ID[4]
            dfs.append(data[(data['Tournament_Name']==t_name) & (data['Player_First_Name']==f_name) & \
                          (data['Player_Last_Name']==l_name) & (data['Round']==ROUND) & (data['Hole']==hole)])
        print y,len(dfs)
    return dfs
```

```python
## let's see what percent of the data consists of holes with additional shots and/or holes with 0 in the coordinates
## before the last shot
def get_size_of_data_without_holes_with_dup_shots ():
    for y in range(2003,2017):
        data = pd.read_csv('data/rawdata/'+str(y)+'.txt', sep = ';') ## read in data
        before = len(data) ##len before
        
        ##processing
        data.columns = np.array([str(i).strip().replace(' ','_') for i in list(data.columns.values)])
    
        data['X_Coordinate'] = [str(i).replace(' ','') for i in data['X_Coordinate']]
        data['Y_Coordinate'] = [str(i).replace(' ','') for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = [str(i).replace(' ','') for i in data['Z_Coordinate']]
        data['X_Coordinate'] = [str(i).replace(',','') for i in data['X_Coordinate']] 
        data['Y_Coordinate'] = [str(i).replace(',','') for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = [str(i).replace(',','') for i in data['Z_Coordinate']]
        data['X_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['X_Coordinate']]
        data['Y_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Z_Coordinate']]
        data['X_Coordinate'] = pd.to_numeric(data['X_Coordinate'])
        data['Y_Coordinate'] = pd.to_numeric(data['Y_Coordinate'])
        data['Z_Coordinate'] = pd.to_numeric(data['Z_Coordinate'])
        
        cols = ['Year','Course_#','Player_#','Round','Hole'] 
        muchacho = set([tuple(i) for i in data[data.Shot>data.Hole_Score][cols].as_matrix().tolist()])
        muchacho.update([tuple(i) for i in data[(data.Shot!=data.Hole_Score) & ((data.X_Coordinate==0) | (data.Y_Coordinate==0) | \
                                             (data.Z_Coordinate==0))][cols].as_matrix().tolist()])
        after =  sum([1 for i in data[cols].as_matrix().tolist() if tuple(i) not in muchacho])
        print after
        prct_lost = (before - after)/float(before)*100
        print 'Year: %d . Data shrunk by %g percent.' % (y,prct_lost)
```

```python
## let's see what percent of the data consists of holes with additional shots and/or holes with 0 in the coordinates
## before the last shot
def eliminate_holes_with_issues ():
    for y in range(2003,2017):
        data = pd.read_csv('data/rawdata/'+str(y)+'.txt', sep = ';') ## read in data
        before = len(data) ##len before
        
        ##processing
        data.columns = np.array([str(i).strip().replace(' ','_') for i in list(data.columns.values)])
    
        data['X_Coordinate'] = [str(i).replace(' ','') for i in data['X_Coordinate']]
        data['Y_Coordinate'] = [str(i).replace(' ','') for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = [str(i).replace(' ','') for i in data['Z_Coordinate']]
        data['X_Coordinate'] = [str(i).replace(',','') for i in data['X_Coordinate']] 
        data['Y_Coordinate'] = [str(i).replace(',','') for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = [str(i).replace(',','') for i in data['Z_Coordinate']]
        data['X_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['X_Coordinate']]
        data['Y_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Y_Coordinate']]
        data['Z_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Z_Coordinate']]
        data['X_Coordinate'] = pd.to_numeric(data['X_Coordinate'])
        data['Y_Coordinate'] = pd.to_numeric(data['Y_Coordinate'])
        data['Z_Coordinate'] = pd.to_numeric(data['Z_Coordinate'])
        
        cols = ['Year','Course_#','Player_#','Round','Hole'] 
        muchacho = set([tuple(i) for i in data[data.Shot>data.Hole_Score][cols].as_matrix().tolist()])
        muchacho.update([tuple(i) for i in data[(data.Shot!=data.Hole_Score) & ((data.X_Coordinate==0) | (data.Y_Coordinate==0) | \
                                             (data.Z_Coordinate==0))][cols].as_matrix().tolist()])
        inds = [u for u,i in enumerate(data[cols].as_matrix().tolist()) if tuple(i) not in muchacho]
        data = data.iloc[inds]
        return data
```
