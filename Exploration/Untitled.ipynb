```python
import pandas as pd
pd.options.display.max_rows = 999
import numpy as np
import warnings
```

```python
data = pd.read_csv('./../data/rawdata/%d.txt' % 2016, sep=';')
data.columns = [col.strip().replace(' ','_') for col in data.columns]
```

```python
data['#_of_Strokes'].value_counts()
```

```python
data['Shot_Type(S/P/D)'].value_counts()
```

```python
cols = ['Year','Tourn.#','Course_#','Round','Hole','Player_#']
```

```python
tups_with_drops = set(tuple(row.values) for row_ind,row in data.loc[data['Shot_Type(S/P/D)']=='D',cols].iterrows())
```

```python
len(tups_with_drops)
```

```python
tups_with_penalties = set(tuple(row.values) for row_ind,row in data.loc[data['Shot_Type(S/P/D)']=='P',cols].iterrows())
```

```python
len(tups_with_penalties)
```

```python
tups_with_provisionals = set(tuple(row.values) for row_ind,row in data.loc[data['Shot_Type(S/P/D)']=='Pr',cols].iterrows())
```

```python
len(tups_with_provisionals)
```

```python
len(tups_with_drops.intersection(tups_with_penalties))
```

```python
len(tups_with_drops.intersection(tups_with_provisionals))
```

```python
len(tups_with_penalties.intersection(tups_with_provisionals))
```

```python
data.groupby('Shot_Type(S/P/D)')['#_of_Strokes'].value_counts()
```

```python
data.loc[([tuple(row.values) in tups_with_provisionals.intersection(tups_with_penalties).intersection(tups_with_drops) 
          for row_ind,row in data[cols].iterrows()]) & (data.Round==4),
         cols+['Shot_Type(S/P/D)','#_of_Strokes',
               'Distance_to_Hole_after_the_Shot','Hole_Score',
               'X_Coordinate','Y_Coordinate','Z_Coordinate']]
```
