Let's see what percent of the data consists of holes with additional shots and/or holes with 0 in the coordinates before the last shot.

```python
data.head()
```

```python
for year in range(2003,2017):
    data = pd.read_csv('./../data/rawdata/%d.txt' % year, sep = ';') ## read in data

    ## making columns usable
    data.columns = [i.strip().replace(' ','_') for i in data.columns]

    ## converting all coords to str to strip out white space
    data['X_Coordinate'] = [str(i).replace(' ','') for i in data['X_Coordinate']]
    data['Y_Coordinate'] = [str(i).replace(' ','') for i in data['Y_Coordinate']]
    data['Z_Coordinate'] = [str(i).replace(' ','') for i in data['Z_Coordinate']]
    
    ## getting rid of commas
    data['X_Coordinate'] = [str(i).replace(',','') for i in data['X_Coordinate']] 
    data['Y_Coordinate'] = [str(i).replace(',','') for i in data['Y_Coordinate']]
    data['Z_Coordinate'] = [str(i).replace(',','') for i in data['Z_Coordinate']]
    
    ## placing negative signs in proper place
    data['X_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['X_Coordinate']] 
    data['Y_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Y_Coordinate']]
    data['Z_Coordinate'] = ['-' + str(i)[:len(str(i))-1] if str(i)[len(str(i))-1]=='-' else i for i in data['Z_Coordinate']]
    
    ## converting to number
    data['X_Coordinate'] = pd.to_numeric(data['X_Coordinate'])
    data['Y_Coordinate'] = pd.to_numeric(data['Y_Coordinate'])
    data['Z_Coordinate'] = pd.to_numeric(data['Z_Coordinate'])

    ## Dropping the match-play championship because there are blank Hole_Scores
    data = data[data['Permanent_Tournament_#']!=470]
    
    ## Ensuring Hole Score and Shot are numerics
    data.Hole_Score = pd.to_numeric(data.Hole_Score)
    data.Shot = pd.to_numeric(data.Shot)

    len_before = len(data)    
    id_cols = ['Year','Permanent_Tournament_#','Player_#','Round','Hole'] ## cols to identify player-hole
    
    ## player-holes that have more shots than the Hole Score
    to_remove = set([tuple(i) for i in data[data.Shot>data.Hole_Score][id_cols].values.tolist()])
    
    ## player-holes that have coordinates that are 0 for shots that are not the last of the hole
    to_remove.update([tuple(i) for i in 
                      data[(data.Shot!=data.Hole_Score) & ((data.X_Coordinate==0) | 
                                                           (data.Y_Coordinate==0) |
                                                           (data.Z_Coordinate==0))][id_cols].values.tolist()])
    
    ## find number of shots that are alright on both conditions
    after =  sum(1 for i in data[id_cols].values.tolist() if tuple(i) not in to_remove)
    pct_lost = (before - after)/float(before)*100
    print 'Year: %d . Data shrunk by %g percent.' % (year,pct_lost)
    print 'Number of fine shots: %d ' % after + '\n'
```
